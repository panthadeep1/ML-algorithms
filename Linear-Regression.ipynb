```python
##Author: panthadeep_b; Time: 12.09 IST
##WAP to implement linear regression

import sys
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
from array import *
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression  
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import datasets

data_set = pd.read_csv("housing.csv");
print("The loaded dataset is:\n", data_set);

##Select the features for model-fit
x = data_set.iloc[:,0];
print("Dim 1:\n", x);

y = data_set.iloc[:,5];
print("Dim 2:\n", y);

##***Convert data to suit the scikit-learn***
x = x.to_numpy().reshape(len(x),1);
y = y.to_numpy().reshape(len(y),1);

##Splitting the dataset into training and test set
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.3, random_state=0);

##Fit-train the model on Train data
model = LinearRegression();
model.fit(x_train,y_train);

##Model predicted value
y_test_pred = model.predict(x_test);
y_train_pred = model.predict(x_train);

#Plot test data
plt.scatter(x_test, y_test, color='green');
plt.title('Test Data');
plt.xlabel('Size');
plt.ylabel('Price');
plt.xticks(());
plt.yticks(());

#Plot test-predictions
plt.plot(x_test, y_test_pred, color='red', linewidth=3);
plt.show();


#Plot train data
plt.scatter(x_train, y_train, color='orange');
plt.title('Train Data');
plt.xlabel('Size');
plt.ylabel('Price');
plt.xticks(());
plt.yticks(());

#Plot train-predictions
plt.plot(x_train, y_train_pred, color='blue', linewidth=3);
plt.show();


##------Accuracy scores-------
rmse_test = math.sqrt(mean_squared_error(y_test,y_test_pred)); ##Test RMSE
rmse_train = math.sqrt(mean_squared_error(y_train,y_train_pred)); ##Train RMSE

print("Train RMSE (error):", rmse_train);
print("Test RMSE (error):", rmse_test);


r2s_test = r2_score(y_test,y_test_pred); ##Test R2-score
r2s_train = r2_score(y_train,y_train_pred); ##Train R2-score

print("Train R2-score (error):", r2s_train);
print("Test R2-score (error):", r2s_test);


```
