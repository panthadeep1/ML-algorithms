```python
##Author: panthadeep_b; Time: 17.09 IST - 15.Nov.25
##WAP to implement Random-forest classifier Source: tpointtech.com

import sys
import numpy as np
import pandas as pd
from array import *
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

print("----Implement the Random-forest ML classifier----");

##Load the data-set
data_set = pd.read_csv("housing.csv");
#print("The loaded dataset is: ", data_set);

##Extracting independent and dependent variables
x = data_set.iloc[:,[0,1,2,3]];
y = data_set.iloc[:, 15];

##***Convert data to suit the scikit-learn***
x = x.to_numpy().reshape(len(x),4);
#x = np.ravel(x);

#y = y.to_numpy().reshape(len(y),1);
y = np.ravel(y);

#print("X is:", x);
#print("Y is:", y);

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0);

#feature scaling
std = StandardScaler();
x_train = std.fit_transform(x_train);
x_test = std.transform(x_test);

##Fitting Decision Tree classifier to the training set
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 10, criterion = "entropy");
model.fit(x_train, y_train);

##Predicting the test set result
y_pred = model.predict(x_test);

##Creating confusion matrix
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
cm = confusion_matrix(y_test, y_pred);

print("The confusion matrix is:", cm);
print("The classification report is:", classification_report(y_test, y_pred));

```
