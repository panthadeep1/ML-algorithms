```python
##Author: panthadeep_b; Time: 19.23 IST - 18.Nov.25
##WAP to implement K-Means clustering algorithm Source: tpointtech.com

import sys
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans

data_set = pd.read_csv('Iris.csv');
#print("The loaded dataset is:", data_set);

##Extracting the independent variables
x = data_set.iloc[:, [2,3]].values;
#***Convert data to suit the scikit-learn***
#x = x.to_numpy().reshape(len(x),2);
#print("X is:", x);

#Using for loop for iterations from 1 to 10.
wcss_list = [];
for i in range(1, 11):  
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state= 42)  
    kmeans.fit(x)  
    wcss_list.append(kmeans.inertia_)  
plt.plot(range(1,11), wcss_list);
plt.title("The Elbow Method Graph");
plt.xlabel('Number of Clusters(K)');
plt.ylabel('wcss_list');
plt.show();
    
```

```python
#Training the K-means model on a dataset
kmeans = KMeans(n_clusters = 3, init='k-means++', random_state=42);
y_predict = kmeans.fit_predict(x);

#visualizing the clusters
plt.scatter(x[y_predict == 0, 0], x[y_predict == 0, 1], s = 100, c = 'blue', label = 'Cluster 1') #for first cluster  
plt.scatter(x[y_predict == 1, 0], x[y_predict == 1, 1], s = 100, c = 'green', label = 'Cluster 2') #for second cluster  
plt.scatter(x[y_predict== 2, 0], x[y_predict == 2, 1], s = 100, c = 'red', label = 'Cluster 3') #for third cluster
#plt.scatter(x[y_predict== 3, 0], x[y_predict == 3, 1], s = 100, c = 'orange', label = 'Cluster 4') #for fourth cluster

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroid') 

plt.title('Iris flowers')  
plt.xlabel('Sample petal')  
plt.ylabel('Flower class')  
plt.legend()  
plt.show()  
```
